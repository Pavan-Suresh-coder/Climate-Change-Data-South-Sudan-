```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
SouthSudan = pd.read_csv("SouthSudanCarbonDioxide.csv")

SouthSudan.plot(
    kind = 'line',
    x = 'Year',
    y = 'CO2 emissions from fuel combustion, South Sudan',
    title = 'South Sudan - Carbon Dioxide Emissions from Fuel Combustion 2000-2023',
    xlabel = 'Years',
    ylabel = 'Carbon Dioxide emission (MTCO2)'
)
plt.show()

SouthSudanTwo = pd.read_csv("CrudeOilSouthSudan.csv")

ax = SouthSudanTwo.plot(
    kind = 'bar',
    x = 'Year',
    y = 'Crude oil production, South Sudan',
    title = 'South Sudan - Domestic Crude Oil Production',
    xlabel = 'Years',
    ylabel = 'Carbon Dioxide emission (MTCO2)'
)

ax.ticklabel_format(style='plain', axis='y')

plt.show()
```

```python
#Energy Production - Different Ways#

#1. Electricity Generation#
SouthSudanThree = pd.read_csv("SouthSudanElectricityGeneration.csv")

SouthSudanThree.plot(
    kind = 'bar',
    x = 'Electricity generation sources, South Sudan, 2023',
    y = 'Value',
    xlabel = 'Type of Electricity',
    ylabel = 'Amount of Energy (GWh)',
    title = 'Amount of Energy Produced South Sudan, 2023'
)
plt.show()

```

```python
#2. Energy Conumption as of 2023 - Trends#
SouthSudanFour = pd.read_csv("SouthSudanConsumptionBySector.csv")

Trends_Sorting = SouthSudanFour[SouthSudanFour['electricity final consumption by sector in South Sudan'] == 'Industry']

Trends_Sorting.plot(kind = 'line', x = 'Year', y = 'Value', xlabel = 'Years', ylabel = 'Amount of Energy Produced (TJ/Terra Joules)', title = 'Terra Joules Energy Produced in Industrial Sector, 2017-2023')
plt.show()

Trends_SortingTwo = SouthSudanFour[SouthSudanFour['electricity final consumption by sector in South Sudan'] == 'Commercial and Public Services']

Trends_SortingTwo.plot(
    kind = 'line',
    x = 'Year',
    y = 'Value',
    xlabel = 'Years',
    ylabel = 'Amount of Energy Produced (TJ/Terra Joules)',
    title = 'Terra Joules Energy Produced in Commercial and Public Services Sector, 2012-2023'
)
plt.show()

Trends_SortingThree = SouthSudanFour[SouthSudanFour['electricity final consumption by sector in South Sudan'] == 'Agriculture and forestry']

Trends_SortingThree.plot(
    kind = 'line',
    x = 'Year',
    y = 'Value',
    xlabel = 'Years',
    ylabel = 'Amount of Energy Produced (TJ / Terra Joules)', 
    title = 'Terra Joules Energy Produced in Agriculture and Forestry Sector, 2012-2023'
)
plt.show()

Trends_SortingFour = SouthSudanFour[SouthSudanFour['electricity final consumption by sector in South Sudan'] == 'Residential']

Trends_SortingFour.plot(
    kind = 'line',
    x = 'Year',
    y = 'Value',
    xlabel = 'Years',
    ylabel = 'Amount of Energy Produced (TJ / Terra Joules)',
    title = 'Terra Joules Energy Produced in Resedential Sector, 2012-2023'
)
plt.show()
```
