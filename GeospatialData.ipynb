```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("5YTDRainfall.csv")

total = df['rfh'].sum()
total2 = df['rfh_avg'].sum()

total_rfh = df[df['adm_level'] == 1]['rfh'].sum() #Adds upp all the elemetns in rfh#
total_rfh2 = df[df['adm_level'] == 1]['rfh_avg'].sum() #Adds up the


plt.bar(['Total Recorded RFH'], [total_rfh])
plt.bar(['Average RFH'], [total_rfh2])
plt.title('Grand Total of All Recorded Rainfall between January 2022 - January 2026')
plt.ylabel('Total Rainfall (mm)')
plt.savefig('grand_total_rfh.png')

RainDifference = total_rfh2 - total_rfh 

print(f"The differece in the amount of rain is {RainDifference:.2f} ml.")
```

```python
df = pd.read_csv("SoilPH.csv")

df_valid = df[(df['PHAQ'] > 0) & (df['PHKC'] > 0)]

means = df_valid[['PHAQ', 'PHKC']].mean()

plt.bar(means.index, means.values, color=['blue', 'red'], alpha=0.7)

# 4. Formatting
plt.title('Average pH: PHAQ vs PHKC (1,489 Samples)')
plt.ylabel('Average pH Value')
plt.ylim(0, 8) # pH scale typically shown 0-14, but zoomed for clarity

for i, v in enumerate(means):
    plt.text(i, v + 0.1, f'{v:.2f}', ha='center', fontweight='bold')

plt.savefig('average_ph_bar_graph.png')
plt.show()
```
