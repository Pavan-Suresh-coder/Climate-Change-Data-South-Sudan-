```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Battle Related deaths in South Sudan#
#The deaths are in terms of 1000#

battle = pd.read_csv("SSbattledeaths.csv")
displaced = pd.read_csv("DisplacedPeople.csv")

battle.plot(
    kind = 'line',
    x = 'Year',
    y = 'Battle-related deaths',
    xlabel = 'Years',
    ylabel = 'Number of deaths (in thousands)',
    title = 'Battle Related Deaths in South Sudan between 2012- 2024'
)
plt.show()

#Although it shows that that the battle related deaths have decreased, it is ranked lowest on countries which to visit.#

#Now to prove this I will analyze the amount of internally displaced people due to crimes#

displaced.plot(
    kind = 'bar',
    x= 'Year',
    y = 'New IDP Displacements',
    xlabel = 'Years',
    ylabel = 'Internally Displaced People (in millions)',
    title = 'Internally Displaced People in South Sudan due to War/Conflict between 2011 - 2023'
)
plt.show()


```

```python
#Analyzing Incidents within South Sudan#
#This has teh specific region as well to help support the info#

incidents = pd.read_csv("IncidentsPerCity.csv")

plt.figure(figsize=(15, 7)) 

ax = incidents.plot(
    kind = 'bar',
    x = 'Admin2',
    y = 'Incident',
    ax = plt.gca(), # Tells pandas to use the figure size we just set
    color = 'skyblue',
    edgecolor = 'black'
)

plt.xticks(rotation=45, ha='right')

# 4. Labels and Title
plt.xlabel('Regions/Cities in South Sudan', fontsize=12)
plt.ylabel('Number of Incidents', fontsize=12)
plt.title('Number of Incidents Per City/Region between January - December 2024', fontsize=14)

# 5. Prevent clipping
plt.tight_layout()
plt.show()
```

```python
#Data of those with acces (% of people with access)#

electricity = pd.read_csv("ElectrictyAccess.csv")

electricity.plot(
    kind = 'line',
    x = 'Year',
    y = 'Access to electricity (% of population)',
    xlabel = 'Year',
    ylabel = '% of population without electricity',
    title = 'Percentage of People without Access to electricity in South Sudan between 2007-2023'
)
plt.show()

print("Note the population of South Sudan is 12.7 million. This indiacates that over 1 million people have no access to basic needs.")
print("\n")

#Now I will compare unemployment to help prove this graph#

unemployment = pd.read_csv("Unemployment.csv")

unemployment.plot(
    kind = 'line',
    x = 'Year',
    y = 'Unemployment, total (% of total labor force)',
    xlabel = 'Year',
    ylabel = 'Unempolyment Rate (% people)',
    title = '% Unemployment in South Sudan between 1991 - 2023'
)
plt.show()
```
